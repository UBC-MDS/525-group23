````python
## You could download it from your bucket, or you can use the file that I have in my bucket. 
## You should be able to access it from my bucket using your key and secret
session = boto3.Session(
    aws_access_key_id="...",
    aws_secret_access_key="...",
)

# 1. Read the data CSV from your s3 bucket.
df = wr.s3.read_csv(path="s3://mds-s3-student61/output/ml_data_SYD.csv", index_col=0, parse_dates=True, boto3_session= session)
df.head()
````

````python
df.shape
````

````python
# 2. Drop rows with nans.
df = df.dropna()
````

````python
# sanity check
df.shape
````

````python
# 3. Split the data into train (80%) and test (20%) portions with random_state=123.
train_df, test_df = train_test_split(df, test_size=0.2, random_state=123)
X_train, y_train = train_df.drop(columns=["Observed"]), train_df["Observed"]
X_test, y_test = test_df.drop(columns=["Observed"]), test_df["Observed"]
````

````python
# 4. Carry out EDA of your choice on the train split.

X_train.info()
````

````python
X_train.describe()
````

````python
# use pandas .plot() function for EDA
train_df["Observed"].plot(kind="hist", bins=200, color = "green", lw=0);
plt.xlabel("Observed rainfall (mm)");
plt.xlim(-1,40);
````

````python
# Use pandas ProfileReport for EDA
profile = ProfileReport(train_df, title="Pandas profiling report", minimal=True);
profile.to_notebook_iframe()
````

````python
profile.to_file("pandas_profining_report_task3.html")
````

````python
# 5. Train ensemble machine learning model using RandomForestRegressor and evaluate with metric of your 
# choice (e.g., RMSE) by considering Observed as the target column.

model = RandomForestRegressor()
model.fit(X_train, y_train)
````

````python
# We chose RMSE as a metric to evaluate our model
print(f"Train RMSE: {mean_squared_error(y_train, model.predict(X_train), squared=False):.2f}")
print(f" Test RMSE: {mean_squared_error(y_test, model.predict(X_test), squared=False):.2f}")
````

````python
# 6. Discuss your results. Are you getting better results with ensemble models compared to the individual climate models?

y_preds, y = model.predict(X_test), pd.DataFrame(y_test)
y['y_preds'] = y_preds

# Use pandas .plot() function to plot the predicted rainfall vs. observed rainfall
y.plot(colormap='Paired');
plt.xlabel("Time");
plt.legend(["Observed rainfall", "Predicted rainfall"]);
````

**Observations:**

As we can see, using the `RandomForestRegressor()` we slightly improved predictions yet when looking at the ground truth of observed rainfall we can tell that the prediction error is quite high.
